```
weather_ww2_feat = weather_ww2[['MaxTemp', 'MinTemp']]
weather_ww2_feat.head()
```

```
len(weather_ww2_feat.MaxTemp)
```

```
plt.scatter(x='MaxTemp', y='MinTemp', data=weather_ww2_feat)
```

```
from sklearn import preprocessing
scaler = preprocessing.RobustScaler()
robust_scaled_df = scaler.fit_transform(weather_ww2_feat)
robust_scaled_df = pd.DataFrame(robust_scaled_df, columns=['MaxTempT', 'MinTempT'])
robust_scaled_df
```

```
import statsmodels.api as sm
import pylab
sm.qqplot(robust_scaled_df['MaxTempT'], line='45') #After RobustScalar
sm.qqplot(robust_scaled_df['MinTempT'], line='45') #After RobustScalar
sm.qqplot(weather_ww2_feat['MaxTemp'], line='45') #Before RobustScalar
sm.qqplot(weather_ww2_feat['MinTemp'], line='45') #Before RobustScalar
pylab.show()
```

```
plt.hist(robust_scaled_df['MaxTempT'])
plt.hist(weather_ww2_feat['MaxTemp'])
```

É possível verificar uma transformação significativa. Usei o Robust Scalar devido aos outliers (como é visivel no histograma laranja). Passaremos para a 2ª fase - treinar o modelo

```
ww2_final = pd.concat([robust_scaled_df, weather_ww2_feat], axis=1, sort=False)
ww2_final.head()
```

```
ww2_final.drop('MinTempT', axis=1, inplace=True)
ww2_final.drop('MaxTemp', axis=1, inplace=True)
ww2_final.head()
```

# Data Training Model

```
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn import metrics
```

```
#x = ww2_final.drop(['Id', 'Species'], axis = 1)
x = ww2_final
y = ww2_final['MinTemp']
```

```
x_train, x_test, y_train, y_test = train_test_split ( x, y , test_size=0.2, random_state=42)
```

```
len(y_test)
```

```
lm = linear_model.LinearRegression()
lm_cvs = cross_val_score(lm, x, y, cv = 5)
print (lm_cvs)
print (lm_cvs.mean())

#normal fit
#lm_model = lm.fit(x_train, y_train)
#predictions = lm.predict(x_test)
```

Usar o RMSE para quantificar o modelo

```
#normal fit
lm = linear_model.LinearRegression()
lm_model = lm.fit(x_train, y_train)
predictions = lm.predict(x_test)
```

```
print(np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
