# Recieving an Array

```
from shapely.geometry import Point, Polygon
array = [-35, -11, 190, 165, 1, 45.7571, 41.2699]
```

```
XR = array[0]
YR = array[1]
XLEN = array[2]
YLEN = array[3]
SIDE = array[4]
XCOMP = array[5]
YCOMP = array[6]
```

```

if SIDE==1:
    area = ['A1', 'A2', 'A3', 'B1', 'B2', 'B3', 'C1', 'C2', 'C3']
else:
    area = ['D1', 'D2', 'D3', 'E1', 'E2', 'E3', 'F1', 'F2', 'F3']
```

# Get Coordinates

```

P1 = [XR, YR]
P2 = [XLEN/3 +(XR), YR]
P3 = [XLEN/3 +(XLEN/3+(XR)), YR]
P4 = [XLEN+XR, YR]
P5 = [XR, YLEN/3+(YR)]
P6 = [XLEN/3+(XR), YLEN/3+(YR)]
P7 = [XLEN/3+(XLEN/3+(XR)), YLEN/3+(YR)]
P8 = [XLEN+XR, YLEN/3+(YR)]
P9 = [XR, YLEN/3+(YLEN/3+(YR))]
P10 = [XLEN/3+XR, YLEN/3+(YLEN/3+(YR))]
P11 = [XLEN/3+(XLEN/3+(XR)), YLEN/3+(YLEN/3+(YR))]
P12 = [XLEN+XR, YLEN/3+(YLEN/3+(YR))]
P13 = [XR, YLEN + YR]
P14 = [XLEN/3 + (XR), YLEN+YR]
P15 = [XLEN/3+(XLEN/3+(XR)), YLEN+YR]
P16 = [XLEN+XR, YLEN+YR]
```

```
areadefinition = ['A1', 'A2', 'A3', 'B1', 'B2', 'B3', 'C1', 'C2', 'C3'] #não associado às áreas. apenas para colocar indicesaa

areadefinition[0] = Polygon([(P13), (P14), (P10), (P9)]) #A1 D1
areadefinition[1] = Polygon([P14, P15, P11, P10]) #A2 D2
areadefinition[2] = Polygon([P15, P16, P12, P11]) #A3 D3
areadefinition[3] = Polygon([P9, P10, P6, P5]) #B1 E1
areadefinition[4] = Polygon([P10, P11, P7, P6]) #B2 E2
areadefinition[5] = Polygon([P11, P12, P8, P7]) #B3 E3 
areadefinition[6] = Polygon([P5, P6, P2, P1]) #C1 F1
areadefinition[7] = Polygon([P6, P7, P3, P2]) #C2 F2
areadefinition[8] = Polygon([P7, P8, P4, P3]) #C3 F3

searchcomp = Point(XCOMP, YCOMP)
```

# Find the Area of the Damaged Component

```
if areadefinition[0].contains(searchcomp):
    print (area[0])
elif areadefinition[1].contains(searchcomp):
    print (area[1])
elif areadefinition[2].contains(searchcomp):
    print (area[2])
elif areadefinition[3].contains(searchcomp):
    print (area[3])
elif areadefinition[4].contains(searchcomp):
    print (area[4])
elif areadefinition[5].contains(searchcomp):
    print (area[5])
elif areadefinition[6].contains(searchcomp):
    print (area[6])
elif areadefinition[7].contains(searchcomp):
    print (area[7])
elif areadefinition[8].contains(searchcomp):
    print (area[8])
```
